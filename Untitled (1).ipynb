```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
adult1=pd.read_csv('adult1.csv')
```

```python
adult1.head()
```

```python
adult1.info()
```

```python
adult1.isnull().sum()
```

```python
print(adult1.isnull().sum())
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
categorical_cols=adult1.select_dtypes(include='object').columns
for col in categorical_cols:
    adult1[col]=le.fit_transform(adult1[col])
```

```python
print(adult1.head())
print(adult1.isnull().sum())
```

```python
plt.figure(figsize=(8,5))
plt.hist(adult1['age'],bins=30,color='lightgreen',edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='income',y='hours.per.week',data=adult1,palette='coolwarm')
plt.title("Hours per week by income")
plt.xlabel("income group")
plt.ylabel("Hours per week")

```

```python
features=['age','education.num','hours.per.week','income']
sns.pairplot(adult1[features],hue='income',palette='Pastel2')
plt.suptitle("pairplot of age, education ,hours,income",y=1.02)
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='income',data=adult1,palette=['#f7b7a3','#a1cfff'])
plt.title("income group distribution",fontsize=14)
plt.xlabel("income group")
plt.ylabel("Count")
plt.show()
```
